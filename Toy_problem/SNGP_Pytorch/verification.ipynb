```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
y_mesh = np.load('y_mesh.npy')
```

```python
plt.figure(figsize=(6, 6))
plt.plot(np.linspace(0, 55, 100), np.linspace(0, 55, 100), '-', color='grey', linewidth = 4, label='Ideal')

for model_name in ['pred_SNGP_4_32.npy', 'pred_SNGP_4_64.npy']: #, 'pred_SNGP_6_32.npy', 'pred_SNGP_6_64.npy']:
    y_hat = np.load(model_name)
    label = 'SNGP_' + model_name.split('_')[2] + '_' + model_name.split('_')[3].split('.')[0]
    plt.scatter(y_mesh, y_hat, s=2, alpha = 0.2, rasterized = True, label = label)
    
plt.xlabel('Ground truth', fontsize=18, fontweight='bold')
plt.ylabel('Prediction', fontsize=18, fontweight='bold')
plt.axis('equal')
plt.xlim([-5, 55])
plt.ylim([-5, 55])
plt.legend(fontsize=16, loc=4)
plt.xticks(fontsize=12, fontweight='bold')
plt.yticks(fontsize=12, fontweight='bold')
plt.tight_layout()
plt.savefig('verification.pdf')
```
